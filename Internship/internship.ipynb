```python
# Perform exploratory data analysis (EDA).

import pandas as pd

# Load the data
df = pd.read_csv('train.csv')

# 1. See the first few rows
print(df.head())

# 2. Get a quick overview (data types, non-null counts)
print(df.info())

# 3. Check shape (rows, columns)
print(df.shape)

# 4. Summary statistics of numeric columns
print(df.describe())

# 5. Check for missing values per column
print(df.isnull().sum())
```

```python
print(df['Sex'].value_counts())
print(df['Embarked'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=df)
plt.show()
```
